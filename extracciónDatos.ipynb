```python
import pandas as pd
import numpy as np 
```

```python
#Cargo mi archivo Inmubles.csv
df = pd.read_csv('Inmuebles.csv')
#df.info()
df
```

```python
#FILTROS BASICOS PARA VARIABLES CUALITATIVAS
#Filtro por objeto 
filtro1=df[df["Operación"] == "Venta"]
filtro1
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("FiltroPorObjeto.csv")
```

```python
#Filtro por filas
#filtro2= df.iloc[300:302,: ]
filtro2=df.iloc[ [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000] , : ]  #filas no consecutivas 
filtro2
```

```python
#Convertir archivo filtrado a CSV
filtro2.to_csv("FiltroFilasNoConsecutivas.csv")
```

```python
#Filtro por columnas
filtro3=df.iloc[ : , 6:9]
#filtro3=df.iloc[ : , [1,4,6]]  #columnas no consecutivas 
filtro3
```
