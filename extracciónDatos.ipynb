```python
import pandas as pd
import numpy as np 
```

```python
#Cargo mi archivo Inmubles.csv
df = pd.read_csv('Inmuebles.csv')
#df.info()
df
```

```python
#FILTROS BASICOS PARA VARIABLES CUALITATIVAS
#Filtro por objeto 
filtro1=df[df["Operación"] == "Venta"]
filtro1
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("FiltroPorObjeto.csv")
```

```python
#Filtro por filas
#filtro2= df.iloc[300:302,: ]
filtro2=df.iloc[ [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000] , : ]  #filas no consecutivas 
filtro2
```

```python
#Convertir archivo filtrado a CSV
filtro2.to_csv("FiltroFilasNoConsecutivas.csv")
```

```python
#Filtro por columnas
filtro3=df.iloc[ : , 6:9]
#filtro3=df.iloc[ : , [1,4,6]]  #columnas no consecutivas 
filtro3
```

```python
#Convertir archivo filtrado a CSV
filtro3.to_csv("FiltroColumnasConsecutivas.csv")
```

```python
#Filtro de filas vs columnas 
filtro4=df.loc[["Lleida","Girona"], ["Vendedor"]]
filtro4
```

```python
#Filtro por cabecera
filtro5=df.head()
filtro5
```

```python
#Convertir archivo filtrado a CSV
filtro5.to_csv("FiltroPorCabecera.csv")
```

```python
#FILTROS BASICOS PARA VARIABLES CUANTITATIVAS
#Filtro por comparacion
filtro6=df[df["Precio Venta"] > 2000000]
filtro6
```

```python
#Convertir archivo filtrado a CSV
filtro6.to_csv("FiltroPorComparación.csv")
```

```python
#FILTROS BASICOS CON CONETORES LOGICOS
#Filtro Y
filtro7=df[(df["Precio Venta"] > 250000) & (df["Operación"] == "Venta")]
filtro7
```

```python
#Convertir archivo filtrado a CSV
filtro7.to_csv("FiltroConectorBasico.csv")
```

```python
#Filtro o
filtro8=df[(df["Precio Venta"] > 250000) | (df["Operación"] == "Venta")]
filtro8
```

```python
#Convertir archivo filtrado a CSV
filtro8.to_csv("Filtro_O.csv")
```

```python
#Filtro not
filtro9=df[~(df["Precio Venta"] >= 500000) & (df["Vendedor"] == "Carmen")]
filtro9
```
